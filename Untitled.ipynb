```python
url_dict = {}
for i in range(len(ground)):
    row = ground.iloc[i]
    url_dict[row['URL']] = list(row)
```

```python
def compare_link(url,url2):
    url = url.strip()
    url2 = url2.strip()
    if url == url2:
        return True
    elif url.split('?')[0] == url2.split('?')[0]:
        return True
```

```python
results = []
with open('new_results.txt') as f:
    for line in f.readlines():
        line = line.strip()
        line = line.replace('[', '').replace(']','').replace("'","")
        results.append(line.split(','))
```
